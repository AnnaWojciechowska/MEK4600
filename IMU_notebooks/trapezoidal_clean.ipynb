```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

lower_limit = 0
upper_limit = 1
bins = 10
x = np.linspace(lower_limit,upper_limit,bins + 1)
y = 2*x**2;
# Trapezoid Rule
y_right = y[1:]
y_left = y[:-1]
dx = np.abs(upper_limit-lower_limit) / bins
area = (dx/2) * np.sum(y_right + y_left)
plt.plot(x,y)
plt.plot(x[1:], y_right, 'ro', alpha = 0.5)
plt.plot(x[:-1], y_left, 'bo', alpha = 0.5)
plt.vlines(x, 0, y, 'lightblue', '--') 
area
```

```python
# same using numpy trapz
%time
A = np.trapz(y,x,dx)
A
```

```python
from scipy import integrate
%time
res = integrate.trapezoid(y,x)
```

```python
from scipy.integrate import cumulative_trapezoid
#note that y and x are swampped
# cumulative has just the sum of all sub areas
%time
cumulative_trapezoid(y, x, initial=0)[-1]
```
