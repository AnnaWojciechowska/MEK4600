```python
from scipy.integrate import quad
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt
```

$\int_0^1 (x^2 + 5) \, dx  = \int_0^1 x^2 \, dx + \int_0^1 5 \, dx$  
 = $ \frac{x^3}{3} + 5x$  
Evaluating this from 0 to 1:  
$ \frac{1}{3} + \frac{15}{3}  -(\frac{0}{3} + 0) = \frac{16}{3}\approx 5 + 0.(3)$


```python
x = np.linspace(0, 1, num=20)
y = x**2 + 5
_ = plt.plot(x,y)
```

```python
%time
def poly_x(x):
    return  x**2 + 5
y_int, error = quad(poly_x, 0, 1)
y_int, error
```

```python
%time
x = np.linspace(0, 1, 100)
y = poly_x(x)
trapz_result = integrate.cumulative_trapezoid(y, x, initial=0)
trapz_result[-1]
```
